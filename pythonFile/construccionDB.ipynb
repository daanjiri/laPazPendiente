```python
import pandas as pd
import json
```

```python
df = pd.read_excel (r'./Lideres sociales asesinados desde 2016.xlsx')
print(df.head())
```

```python
newdf=df.fillna(" ")
print(newdf.head())
```

```python
db=[]
for i in range(len(newdf)) : 
  element={
      "location":newdf.loc[i, "id"],
      "url":newdf.loc[i, "Fuentes"]
  }
  db.append(element)


  print(db) 
```

```python
my_details = {
    'name': 'John Doe',
    'age': 29
}

with open('personal.json', 'w') as json_file:
    json.dump(db, json_file)
```
